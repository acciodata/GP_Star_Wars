```python
# Reading the data
import pandas as pd
import numpy

star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")
```

```python
# Displaying the first 10 rows

star_wars.head(10)
```

```python
# Displaying the column names

star_wars.columns
```

```python
# Generic function for the column names
import re

questions = star_wars.columns.tolist()

def get_col_index(regex, column_names = questions):
    for i, column in enumerate(column_names):
        if re.search(regex, column) is not None:
            col_index_list = i
            print(i)
    col_index = column_names[col_index_list]
    return col_index
```

```python
# Cleaning Step 1: Removing rows with NaN values for RespondentID

star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
# Cleaning Step 2A
# Converting Yes and No answers to True and False respectively.

# Creating a dictionary for the replacement
yes_no = {"Yes": True, "No": False}

# For the column: Have you seen any of the 6 films in the Star Wars franchise?

column_title = get_col_index("6 films")
star_wars[column_title] = star_wars[column_title].map(yes_no)

star_wars.head()
```

```python
# Cleaning Step 2B
# For the column Do you consider yourself to be a fan of the Star Wars film franchise?

column_title = get_col_index("fan of the Star Wars")
star_wars[column_title] = star_wars[column_title].map(yes_no)

star_wars.head()
```

```python
# Cleaning Step 3 - Part 1

# Creating the dictionary for replacement
movie_dictionary = {}

for i in range(3, 9):
    temp_list = star_wars[star_wars.columns[i]].unique().tolist()
    movie_dictionary[temp_list[0]] = True
movie_dictionary["NaN"] = False

# Replacing responses with True or False

for i in range(3, 9):
    star_wars[star_wars.columns[i]] = star_wars[star_wars.columns[i]].map(movie_dictionary)

star_wars.head()    
```

```python
# Cleaning Step 3 - Part 2
# Renaming the column names

new_names_seen = ["seen_1", "seen_2", "seen_3", "seen_4", "seen_5", "seen_6"]

for i in range(3, 9):
    star_wars = star_wars.rename(columns = {star_wars.columns[i]: new_names_seen[i-3]})

star_wars.head()
```

```python
# Cleaning Step 4 - Part 1
# Converting to float

def convert_to_float(df):
    try:
        converted = float(df)
    except Exception as exc:
        converted = None
    return converted

for i in range (9, 15):
     star_wars[star_wars.columns[i]] = star_wars[star_wars.columns[i]].apply(convert_to_float)

star_wars.head()
```

```python
# Cleaning Step 4 - Part 2
# Renaming columns

new_names_ranking = ["ranking_1", "ranking_2", "ranking_3", "ranking_4", "ranking_5", "ranking_6"]

for i in range(9, 15):
    star_wars = star_wars.rename(columns = {star_wars.columns[i]:new_names_ranking[i-9]})

star_wars.head()
```

```python
# Analysis - Finding the highest ranked movie

%matplotlib inline
import matplotlib.pyplot as plt

ranking_mean = star_wars[star_wars.columns[9:15]].mean()

ranking_mean.plot(kind = "bar")
```

Each respondent was asked to rank the movies from 1 to 6 with 1 being the most favorable while 6 being the least favorable. A bar plot of the mean was also displayed, showing ranking_5 to have the smallest ranking of all. This indicates that more people favored The Empire Strikes Back among the 6 movie franchises.

It can also be noticed that the older franchises (ranking_4, ranking_5, and ranking_6) are more popular than the newer franchises (ranking_1, ranking_2, and ranking_3). And among the newer franchises, there seems to be a trend of a newer franchise being less popular than its predecessor.

```python
# Analysis 3A: Part 1 - Dividing the data into star wars fans
# and not star wars fans

# Dataframe for star wars fans responses
star_wars_fan = star_wars[star_wars[star_wars.columns[2]] == True]

# Dataframe for non star wars fans responses
star_wars_not_fan = star_wars[star_wars[star_wars.columns[2]] != True]

# Finding the mean ranking per film
mean_ranking_sw_fan = star_wars_fan[star_wars_fan.columns[9:15]].mean().tolist()
mean_ranking_sw_notfan = star_wars_not_fan[star_wars_not_fan.columns[9:15]].mean().tolist()

# Preparation for Grouped Bar Plots
new_column_names = ["franchise", "fan_response", "non_fan_response"]
bundled_data_list = [new_names_ranking, mean_ranking_sw_fan, mean_ranking_sw_notfan]

# Creating a general function

def df_for_group_plot(data_list, column_names):
    temp_dictionary = {}
    for i, title in enumerate(column_names):
        temp_dictionary[title] = data_list[i]
    new_df = pd.DataFrame(temp_dictionary)
    return (new_df)

fan_ranking_dataframe = df_for_group_plot(bundled_data_list, new_column_names)    
    
fan_ranking_dataframe.plot(x = "franchise", kind = "bar", legend = False)
```

Ranking by Star Wars fans is in blue, while ranking by non-Star-Wars-fans is in orange.

Fan or no fan of the franchise, the best ranked film is still The Empire Strikes Back. 

Noticeably, fans of the franchise ranked the older films better than the newer ones, where A New Hope and Return of the Jedi almost ranking the same. For non-fans of the franchise, Phantom Menace was noticeably ranked better than Revenge of the Sith. 

For both fans and non-fans, the downward trend in preference for the newer franchise is still evident in the plots given above.

```python
# Analysis 3A - Part 2

# Finding the total number of viewers per franchise
total_viewers_sw_fan = star_wars_fan[star_wars_fan.columns[3:9]].sum().tolist()
total_viewers_sw_notfan = star_wars_not_fan[star_wars_not_fan.columns[3:9]].sum().tolist()

# Preparation for Grouped Bar Plots
bundled_data_list = [new_names_seen, total_viewers_sw_fan, total_viewers_sw_notfan]

viewer_dataframe = df_for_group_plot(bundled_data_list, new_column_names) 
viewer_dataframe.plot(x = "franchise", kind = "bar", legend = False)
```

Blue bars represent the number of viewers per franchise for star wars fans while orange bars represent the number of viewers per franchise for non-star-wars-fans.

Not surprisingly, there is a significant higher number of viewers for star wars fans versus those who are not fans.

But the number of viewers for Empire Strikes Back (seen_5) are still greater for all franchises, whether the viewers are fans or not fans of Star Wars. The least watched of all the franchises is still Revenge of the Sith.

```python
# Data Cleaning Next Step - Clean Columns 15 to 29

# Renaming Columns
old_column_names = star_wars.columns[15:29].tolist()
character_names = ["Han Solo", "Luke Skywalker", "Princess Leia Organa", "Anakin Skywalker", "Obi Wan Kenobi", "Emperor Palpatine", "Darth Vader", "Lando Calrissian", "Boba Fett", "C-3P0", "R2 D2", "Jar Jar Binks", "Padme Amidala", "Yoda"]

rename_dictionary = {}

for i in range(0, len(old_column_names)):
    rename_dictionary[old_column_names[i]] = character_names[i]

star_wars = star_wars.rename(columns = rename_dictionary)
    
# Creating the numerical equivalent dictionary for the character ratings
old_rating_list = star_wars[star_wars.columns[15]].unique().tolist()
old_rating_list.remove("Han Solo")
old_rating_list.remove(old_rating_list[1])

new_rating_list = [1, 2, 3, 4, 5, 6]

replacement_dictionary = {}
for i in range(0, len(old_rating_list)):
    replacement_dictionary[old_rating_list[i]] = new_rating_list[i]

for character in character_names:
    replacement_dictionary[character] = None

replacement_dictionary["nan"] = None

for i in range(15, 29):
    star_wars[star_wars.columns[i]] = star_wars[star_wars.columns[i]].map(replacement_dictionary)
    
star_wars.head()
```

```python
# Analysis Next Step 4 - Finding the character most liked by the respondents and the character most disliked by the respondents

# Finding the mean rating per character
percharacter_mean_rating = star_wars[star_wars.columns[15:29]].mean()
percharacter_mean_rating.plot(kind = "bar")
```
